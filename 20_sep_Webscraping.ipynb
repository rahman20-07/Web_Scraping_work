```python
#to fetch all judges name & append them to judges list

tags=soup.find_all('strong',{'class':'judge-name'})
judges=[]
for tag in tags:
    judge_name = tag.get_text()
    judges.append(judge_name)

#creating empty list for DOB,DOA & DOR
dob=[]
doa=[]
dor=[]

#inserting dob of chief justice to dob list
tag=soup.find('span',{'class':'judge-dob'})
dob.insert(0,tag.get_text()[-10:])

#inserting doa of chief-justice to doa list
tag=soup.find('div',{'class':'chief-justice-right'})
text=tag.get_text()

i=text.index("(DoA)")
doa.insert(0,text[i+6:i+16])

#inserting dor of chief-justice to dor list
i=text.index("(DoR")
dor.insert(0,text[i+6:i+16])

#appending dob , doa , dor of justices
tags=soup.find_all('p')
text =''
for tag in tags[10:]:
    text=text+tag.get_text()
    txt=tag.get_text()

#import re ---->  coz already imported esliye comment krdi
dates=re.findall("[0-9][0-9]-[01][0-9]-[0-9]{4}",text)

for i in range(0,len(dates)-2,3):
    li=dates[i:i+3]
    dob.append(li[0])
    doa.append(li[1])
    dor.append(li[2])

#creating DataFrame
df=pd.DataFrame()
df['Judge Name']=judges
df['DOB']=dob
df['DOA']=doa
df['DOR']=dor

#print dataframe
df

#storing dataframe as excel file to my device @ specifig location

#df.to_excel(r'D:\Ducat_Course\Statistics\Dataset_Stats\Judges_details.xlsx',index=False)
```
